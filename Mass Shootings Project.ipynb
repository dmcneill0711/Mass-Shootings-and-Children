```python
# Upload 2022 Mass Shootings Table From Wikipedia Website
url = 'https://en.wikipedia.org/wiki/List_of_mass_shootings_in_the_United_States_in_2022'

tables = pd.read_html(url)

tables
```

```python
# Select Relevant Table
Mass_Shootings = pd.DataFrame(tables[0])

# Rename Column
Mass_Shootings = Mass_Shootings.rename(columns = {'2022 date': 'Incident Date'})

# Format The Date To mm/dd/yyyy
Mass_Shootings['Incident Date'] = Mass_Shootings['Incident Date'].str.strip().str.replace(' ','')
Mass_Shootings['Incident Date'] = Mass_Shootings['Incident Date'].str.strip().str.replace('January','01/')
Mass_Shootings['Incident Date'] = Mass_Shootings['Incident Date'].str.strip().str.replace('February','02/')
Mass_Shootings['Incident Date'] = Mass_Shootings['Incident Date'].str.strip().str.replace('March','03/')
Mass_Shootings['Incident Date'] = Mass_Shootings['Incident Date'].str.strip().str.replace('April','04/')
Mass_Shootings['Incident Date'] = Mass_Shootings['Incident Date'].str.strip().str.replace('May','05/')
Mass_Shootings['Incident Date'] = Mass_Shootings['Incident Date'].str.strip().str.replace('June','06/')
Mass_Shootings['Incident Date'] = Mass_Shootings['Incident Date'].str.strip().str.replace('July','07/')
Mass_Shootings['Incident Date'] = Mass_Shootings['Incident Date'].str.strip().str.replace('August','08/')
Mass_Shootings['Incident Date'] = Mass_Shootings['Incident Date'].str.strip().str.replace('September','09/')
Mass_Shootings['Incident Date'] = Mass_Shootings['Incident Date'].str.strip().str.replace('October','10/')
Mass_Shootings['Incident Date'] = Mass_Shootings['Incident Date'].str.strip().str.replace('November','11/')
Mass_Shootings['Incident Date'] = Mass_Shootings['Incident Date'].str.strip().str.replace('December','12/')
Mass_Shootings['Incident Date'] = Mass_Shootings['Incident Date'].str.strip().str.replace(',','/')

# Add Year To End of Date Column To Match Format Of Children_2022 DataFrame
Mass_Shootings['Incident Date'] = Mass_Shootings['Incident Date'].astype(str) + '/22'
Mass_Shootings['Incident Date'] = pd.to_datetime(Mass_Shootings['Incident Date'], errors='coerce')


Mass_Shootings_Renamed = Mass_Shootings.rename(columns={"Incident Date": "incident_date", "City": "city", "State": "state",
                                                        "Dead": "dead", "Injured": "injured", "Total": "total",
                                                       "Description": "description"})

#Final Dataframe
Mass_Shootings_Renamed
```
